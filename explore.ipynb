```python
import matplotlib.pyplot as plt
import numpy as np

# Data
metrics = ['Exact Match', 'F1 Score', 'Has Answer Exact Match', 'Has Answer F1 Score',
           'No Answer Exact Match', 'No Answer F1 Score', 'Best Exact Match', 'Best F1 Score']
n_metrics = len(metrics)

BERT_scores = [27.878, 32.130, 50.405, 58.920, 5.416, 5.416, 50.114, 50.114]
DistilBERT_scores = [23.347, 26.870, 38.630, 45.686, 8.108, 8.108, 50.114, 50.114]
RoBERTa_scores = [37.404, 40.428, 72.588, 78.644, 2.321, 2.321, 50.097, 50.097]

# Create figure and axes
fig, ax = plt.subplots(figsize=(15, 10))

# Set position of bar on X axis
ind = np.arange(n_metrics)
width = 0.25  # the width of the bars

# Create the bars
bar1 = ax.bar(ind - width, BERT_scores, width, label='BERT')
bar2 = ax.bar(ind, DistilBERT_scores, width, label='DistilBERT')
bar3 = ax.bar(ind + width, RoBERTa_scores, width, label='RoBERTa')

# Set the labels, ticks, and title
ax.set_ylabel('Scores')
ax.set_title('Fine-tuned Model Comparison', fontsize=26, weight='bold')
ax.set_xticks(ind)
ax.set_xticklabels(metrics, rotation=45, ha="right")
ax.legend()

# Set the background color of the figure to transparent
fig.patch.set_alpha(0)

# Function to attach a text label above each bar in *rects*, displaying its height
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

# Call the function to add the labels
autolabel(bar1)
autolabel(bar2)
autolabel(bar3)

# Make layout adjustments to accommodate the labels and prevent overlap
plt.tight_layout()

# Save the figure
plt.savefig('images/model_comparison_chart.png', transparent=True)
```
